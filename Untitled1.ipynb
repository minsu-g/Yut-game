```python
import numpy as np

class Yut:
    def __init__(self):
        self.yut_myturn = []
        self.throw_result = None

        self.y_do = '''
┌---┐┌---┐┌---┐┌---┐
| x || x || x ||   |
|   ||   ||   ||   |
| x || x || x ||   |
|   ||   ||   ||   |
| x || x || x ||   |
└---┘└---┘└---┘└---┘
'''
        self.y_gae = '''
┌---┐┌---┐┌---┐┌---┐
| x || x ||   ||   |
|   ||   ||   ||   |
| x || x ||   ||   |
|   ||   ||   ||   |
| x || x ||   ||   |
└---┘└---┘└---┘└---┘
'''
        self.y_gul = '''
┌---┐┌---┐┌---┐┌---┐
| x ||   ||   ||   |
|   ||   ||   ||   |
| x ||   ||   ||   |
|   ||   ||   ||   |
| x ||   ||   ||   |
└---┘└---┘└---┘└---┘
'''
        self.y_yut = '''
┌---┐┌---┐┌---┐┌---┐
|   ||   ||   ||   |
|   ||   ||   ||   |
|   ||   ||   ||   |
|   ||   ||   ||   |
|   ||   ||   ||   |
└---┘└---┘└---┘└---┘
'''
        self.y_mo = '''
┌---┐┌---┐┌---┐┌---┐
| x || x || x || x |
|   ||   ||   ||   |
| x || x || x || x |
|   ||   ||   ||   |
| x || x || x || x |
└---┘└---┘└---┘└---┘
'''
        self.y_backdo = '''
┌---┐┌---┐┌---┐┌---┐
|   || x || x || x |
|   ||   ||   ||   |
| * || x || x || x |
|   ||   ||   ||   |
|   || x || x || x |
└---┘└---┘└---┘└---┘
'''

    def throw(self):
        yut_choice = ['도', '개', '걸', '윷', '모', '빽도']
        yut_prob = [0.1875, 0.375, 0.25, 0.0625, 0.0625, 0.0625]

        self.throw_result = np.random.choice(yut_choice, p=yut_prob)
        self.yut_myturn.append(self.throw_result)
        print(f'윷을 던져 {self.throw_result}가 나왔습니다.')

        if self.throw_result == '도':
            print(self.y_do)
        elif self.throw_result == '개':
            print(self.y_gae)
        elif self.throw_result == '걸':
            print(self.y_gul)
        elif self.throw_result == '윷':
            print(self.y_yut)
        elif self.throw_result == '모':
            print(self.y_mo)
        elif self.throw_result == '빽도':
            print(self.y_backdo)

        return self.throw_result

    def yut_list(self):
        return self.yut_myturn

    def reset_yut(self):
        self.yut_myturn = []
        self.throw_result = None


class Mal(Yut):
    def __init__(self, player, mal):
        super().__init__()
        self.player = player
        self.mal = mal
        self.position = None
        self.stacked = False
        self.stack_mal = mal
        self.current_route = None
        self.is_stack_leader = True
        self.route_name = 'default'

    def mal_move_value(self, yut_result):
        if yut_result == '도':
            return 1
        elif yut_result == '개':
            return 2
        elif yut_result == '걸':
            return 3
        elif yut_result == '윷':
            return 4
        elif yut_result == '모':
            return 5
        elif yut_result == '빽도':
            return -1
        return 0

    def select_route(self, yutpan):
        default_route = yutpan.routes['default']
        from_5 = [5,8,12,14,16,20,23]
        from_0 = [0,7,11,14,17,21,28]
        from_14 = [14,17,21,28]

        pos = self.position

        def in_route(p, r):
            return p in r

        if pos is None:
            self.route_name = 'default'
            self.current_route = default_route
            return

        # 교차점 처리
        if pos == 5:
            self.route_name = 'from_5'
            self.current_route = from_5
            return
        elif pos == 0:
            self.route_name = 'from_0'
            self.current_route = from_0
            return
        elif pos == 14:
            self.route_name = 'from_14'
            self.current_route = from_14
            return

        if self.route_name == 'default':
            route = default_route
        elif self.route_name == 'from_5':
            route = from_5
        elif self.route_name == 'from_0':
            route = from_0
        elif self.route_name == 'from_14':
            route = from_14
        else:
            route = default_route

        if not in_route(pos, route):
            if in_route(pos, from_5):
                self.route_name = 'from_5'
                route = from_5
            elif in_route(pos, from_0):
                self.route_name = 'from_0'
                route = from_0
            elif in_route(pos, from_14):
                self.route_name = 'from_14'
                route = from_14
            elif in_route(pos, default_route):
                self.route_name = 'default'
                route = default_route
            else:
                self.route_name = 'default'
                route = default_route

        self.current_route = route

    def move(self, steps, yutpan):
        if self.position is None:
            self.position = yutpan.start_index
            self.route_name = 'default'
            self.current_route = yutpan.routes['default']

        self.select_route(yutpan)
        route = self.current_route

        current_idx = route.index(self.position)
        new_idx = current_idx + steps

        last_idx = len(route) - 1

        if new_idx > last_idx:
            # 골인
            self.position = None
            return True

        if new_idx < 0:
            if self.route_name == 'default':
                new_idx = 0
            else:
                new_idx = max(0, new_idx)

        self.position = route[new_idx]
        return False

    def mal_catch(self, other_mal_list, game):
        caught = False
        for om in other_mal_list:
            if om.position == self.position and om.position is not None and om.player != self.player:
                print('말을 잡았습니다!')
                caught = True
                if om.stacked:
                    self.unstack_captured(om, game)
                else:
                    om.position = None
                    om.current_route = None
        return caught

    def mal_stack(self, own_mal_list):
        for om in own_mal_list:
            if om != self and om.player == self.player and om.position == self.position and om.position is not None:
                self.stacked = True
                om.stacked = True
                self.stack_mal = self.mal.upper()
                om.position = None
                om.is_stack_leader = False
                om.stacked = False
                om.stack_mal = om.mal
                return True
        return False

    def unstack_captured(self, om, game):
        player = om.player
        mals = game.players_mal_dict[player]
        for m in mals:
            m.position = None
            m.stacked = False
            m.is_stack_leader = True
            m.stack_mal = m.mal

    def print_symbol(self):
        if self.stacked and self.is_stack_leader:
            return self.stack_mal
        elif not self.stacked:
            return self.mal
        else:
            return None


class Yutpan:
    def __init__(self):
        self.board_matrix = [
            ["□", "□", "□", " ", "□", "□", "□"],
            ["□", "□", " ", " ", " ", "□", "□"],
            ["□", " ", "□", " ", "□", " ", "□"],
            [" ", " ", " ", "□", " ", " ", " "],
            ["□", " ", "□", " ", "□", " ", "□"],
            ["□", "□", " ", " ", " ", "□", "□"],
            ["□", "□", "□", " ", "□", "□", "□"]
        ]
        self.rows = 7
        self.cols = 7

        self.squares = []
        for r in range(self.rows):
            for c in range(self.cols):
                if self.board_matrix[r][c] == "□":
                    self.squares.append((r,c))

        self.start_index = self.squares.index((6,6))
        self.routes = {
            'default': [28,22,18,13,9,5,4,3,2,1,0,6,10,15,19,23,24,25,26,27,28]
        }

    def print_board(self, player_mals):
        temp_board = [row[:] for row in self.board_matrix]

        for p, mals in player_mals.items():
            for m in mals:
                if m.position is not None and m.is_stack_leader:
                    r,c = self.squares[m.position]
                    symbol = m.print_symbol()
                    if symbol is not None:
                        temp_board[r][c] = symbol

        print('\n현재 윷판 상태:')
        for row in temp_board:
            print(' '.join(row))
        print()

    def is_all_goal(self, mals):
        return all(m.position is None for m in mals)


class Gamestart:
    def __init__(self):
        self.player1 = input('1번 플레이어의 이름을 입력하세요 : ').strip()
        self.player2 = input('2번 플레이어의 이름을 입력하세요 : ').strip()
        self.players_info = {}
        self.players_mal_dict = {}
        self.current_player = self.player1

    def set_mal(self):
        while True:
            try:
                mal1 = input('{}님이 사용할 말을 소문자 영어 한 글자로 입력하세요 : '.format(self.player1)).strip()
                if len(mal1) != 1 or not mal1.isalpha() or not mal1.islower():
                    raise ValueError('말은 소문자 영어 한 글자로 입력해야합니다.')
                self.players_info[self.player1] = [mal1, mal1]

                mal2 = input('{}님이 사용할 말을 소문자 영어 한 글자로 입력하세요 : '.format(self.player2)).strip()
                if len(mal2) != 1 or not mal2.isalpha() or not mal2.islower():
                    raise ValueError('말은 소문자 영어 한 글자로 입력해야합니다.')
                self.players_info[self.player2] = [mal2, mal2]
                break
            except ValueError as e:
                print(e)

        self.players_mal_dict[self.player1] = [Mal(self.player1, self.players_info[self.player1][0]),
                                               Mal(self.player1, self.players_info[self.player1][1])]
        self.players_mal_dict[self.player2] = [Mal(self.player2, self.players_info[self.player2][0]),
                                               Mal(self.player2, self.players_info[self.player2][1])]

        return self.players_info

    def switch_player(self):
        self.current_player = self.player1 if self.current_player == self.player2 else self.player2


class Myturn:
    def __init__(self, game, yutpan):
        self.game = game
        self.yutpan = yutpan
        self.yut = Yut()

    def prompt_throw(self):
        tries = 0
        while True:
            user_input = input('Enter 키로 윷을 던지세요! : ')
            try:
                if user_input != '':
                    raise ValueError('Enter 키가 아닌 다른 키를 입력했습니다!')
                return True
            except ValueError as e:
                print(e)
                tries += 1
                if tries == 1:
                    print('한 번 더 잘못 던지면 낙입니다!')
                else:
                    print('낙입니다. 다음 차례로 넘어갑니다.')
                    return False

    def playmyturn(self):
        self.yut.reset_yut()

        if not self.prompt_throw():
            return None
        result = self.yut.throw()
        if result is None:
            return None

        while result in ['윷', '모']:
            print("윷/모 나왔으니 한 번 더 던지세요!")
            if not self.prompt_throw():
                return self.yut.yut_list()
            result = self.yut.throw()
            if result is None:
                return None

        return self.yut.yut_list()

    def extra_throw(self):
        self.yut.reset_yut()
        if not self.prompt_throw():
            return None
        result = self.yut.throw()
        if result is None:
            return None
        while result in ['윷', '모']:
            print("윷/모 나왔으니 한 번 더 던지세요!")
            if not self.prompt_throw():
                return self.yut.yut_list()
            result = self.yut.throw()
            if result is None:
                return None
        return self.yut.yut_list()

    def choose_result(self, results):
        while True:
            print("현재 사용 가능한 윷 결과:", results)
            choice = input("사용할 윷 결과를 입력하세요(도/개/걸/윷/모/빽도): ").strip()
            if choice in results:
                results.remove(choice)
                return choice
            else:
                print("해당 윷 결과는 사용 가능한 윷 결과 목록에 없습니다. 다시 입력하세요.")

    def turn_process(self):
        initial_results = self.playmyturn()
        if initial_results is None or len(initial_results) == 0:
            return

        results = initial_results[:]
        current_player = self.game.current_player
        current_mals = self.game.players_mal_dict[current_player]
        other_player = self.game.player1 if current_player == self.game.player2 else self.game.player2
        other_mals = self.game.players_mal_dict[other_player]

        while len(results) > 0:
            selected_result = self.choose_result(results)

            # 말 상태 파악
            stack_leader = [m for m in current_mals if m.position is not None and m.stacked and m.is_stack_leader]
            alive_mals = [m for m in current_mals if (m.position is not None) or (m.position is None and not m.stacked and m.is_stack_leader)]

            if stack_leader:
                # 업힌 말 있으면 자동 선택
                selected_mal = stack_leader[0]
            else:
                # 업힌 말 없으면 말 수 체크
                if len(alive_mals) == 1:
                    # 말 하나만 남았으면 자동 선택
                    selected_mal = alive_mals[0]
                else:
                    if selected_result == '빽도':
                        selectable = [m for m in current_mals if m.position is not None]
                        if len(selectable) == 0:
                            print('이동 가능한 말이 없습니다. 다음 차례로 넘어갑니다.')
                            return
                        print("현재 플레이어 말 상태(빽도 - 보드 위 말만 선택 가능):")
                        for idx, mm in enumerate(current_mals):
                            if mm in selectable:
                                pos_str = f"인덱스 {mm.position}"
                                print(f"{idx+1}. {mm.mal} ({pos_str})")
                        while True:
                            try:
                                choice = input("이동시킬 말 번호를 입력하세요(빽도): ")
                                if not choice.isdigit():
                                    raise ValueError("숫자를 입력하세요.")
                                c_idx = int(choice)-1
                                if c_idx < 0 or c_idx >= len(current_mals):
                                    raise ValueError("잘못된 말 번호.")
                                selected_mal = current_mals[c_idx]
                                if selected_mal not in selectable:
                                    raise ValueError("빽도는 보드 위에 있는 말만 선택할 수 있습니다.")
                                break
                            except ValueError as e:
                                print(e)
                    else:
                        selectable = [m for m in current_mals if m.position is not None]
                        not_started = [m for m in current_mals if m.position is None]

                        if len(selectable) == 0 and len(not_started) == 1:
                            # 시작전 말만 하나 남았다면 자동 선택
                            selected_mal = not_started[0]
                        else:
                            print("현재 플레이어 말 상태:")
                            for idx, mm in enumerate(current_mals):
                                pos_str = "시작 전" if mm.position is None else f"인덱스 {mm.position}"
                                print(f"{idx+1}. {mm.mal} ({pos_str})")
                            while True:
                                try:
                                    choice = input("이동시킬 말 번호를 입력하세요: ")
                                    if not choice.isdigit():
                                        raise ValueError("숫자를 입력하세요.")
                                    c_idx = int(choice)-1
                                    if c_idx < 0 or c_idx >= len(current_mals):
                                        raise ValueError("잘못된 말 번호.")
                                    selected_mal = current_mals[c_idx]
                                    break
                                except ValueError as e:
                                    print(e)

            steps = selected_mal.mal_move_value(selected_result)

            if selected_result == '빽도' and all(m.position is None for m in current_mals):
                print('이동 가능한 말이 없습니다. 다음 차례로 넘어갑니다.')
                return

            is_goal = selected_mal.move(steps, self.yutpan)
            if is_goal:
                # 골인 처리
                pass

            caught = selected_mal.mal_catch(other_mals, self.game)
            if caught:
                extra_res = self.extra_throw()
                if extra_res is None:
                    return
                else:
                    results.extend(extra_res)

            selected_mal.mal_stack(current_mals)
            self.yutpan.print_board(self.game.players_mal_dict)

            if self.yutpan.is_all_goal(current_mals):
                print(f"{current_player}님이 모든 말을 골인시켰습니다! 승리!")
                return "WIN"

        return "CONTINUE"


def main():
    game = Gamestart()
    game.set_mal()
    yutpan = Yutpan()

    print('윷놀이를 시작합니다!')

    while True:
        print(f"{game.current_player}님의 차례입니다.")
        yutpan.print_board(game.players_mal_dict)

        turn = Myturn(game, yutpan)
        result = turn.turn_process()

        if result == "WIN":
            break
        elif result in [None, "CONTINUE"]:
            game.switch_player()

if __name__ == "__main__":
    main()
```
